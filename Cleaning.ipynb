## Data Analysis
### A little overview of the dataset

```python
def display_counting(df,x):
    return df[[x,'Name']].groupby(by=[x]).count().sort_values(by=['Name'],ascending=False)
```

```python
display_counting(df,'Year')
```

```python
df=df[df['Year']!=2020]
df=df[df['Year']!=2017]
```

```python
display_counting(df,'Genre')
```

```python
display_counting(df,"Platform")
```

```python
display_counting(df,"Publisher").head(10)
```

```python
df[['Platform','Global_Sales']].groupby(by=['Platform']).mean().sort_values(by=['Global_Sales'],ascending=False)
```

## Data Cleaning Pt.2
### Replacing the commas in decimal numbers with dots (It's important so PowerBI will read it as a decimal number)

```python
def replace_commas(df,x):
    df[x]=df[x].apply(lambda b : str(b).replace('.',','))
```

```python
df.columns
```

```python
replace_commas(df,'Global_Sales')
```

```python
replace_commas(df,df.columns[-2])
```

```python
replace_commas(df,df.columns[-3])
```

```python
replace_commas(df,df.columns[-4])
```
